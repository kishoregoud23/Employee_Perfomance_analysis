# Feature Overview

The dataset comprises 1200 records with 28 features, including both categorical and numerical variables. Below is a brief description of each feature:


* Categorical Features: 

1. Gender: Employee's gender.
2. EducationBackground: Field of education (e.g., Life Sciences, Medical).
3. MaritalStatus: Marital status of the employee.
4. EmpDepartment: Department in which the employee works.
5. EmpJobRole: Specific job role/title of the employee.
6. BusinessTravelFrequency: Frequency of business travel.
7. OverTime: Indicates if the employee works overtime.
8. EducationLevel: Highest level of education attained.
9. EmpEnvironmentSatisfaction: Satisfaction level with the work environment.
10. EmpJobLevel: Job level within the organization.
11. EmpJobSatisfaction: Satisfaction with the job role.
12. EmpJobInvolvement: Level of involvement in the job.
13. EmpRelationshipSatisfaction: Satisfaction with workplace relationships.
14. WorkLifeBalance: Balance between work and personal life.
15. Attrition: Indicates if the employee has left the company.

* Numerical Features:

1. Age: Age of the employee.
2. DistanceFromHome: Distance between home and workplace.
3. EmpHourlyRate: Hourly wage rate.
4. NumCompaniesWorked: Number of previous companies worked for.
5. EmpLastSalaryHikePercent: Percentage of the last salary hike.
6. TotalWorkExperienceInYears: Total years of work experience.
7. TrainingTimesLastYear: Number of training sessions attended last year.
8. ExperienceYearsAtThisCompany: Years of experience at INX Future Inc.
9. ExperienceYearsInCurrentRole: Years in the current role.
10. YearsSinceLastPromotion: Years since the last promotion.
11. YearsWithCurrManager: Years working with the current manager.

* Target Variable:
PerformanceRating: Performance rating assigned to the employee.

-- Will use Hyperparameter Tunning in all algorithms to get the best accuracy
- **Hyperparameter Tunning** is the process of finding the best set of hyperparameters for a machine learning model to achieve optimal performance on a given dataset.
- In hyperparameter tunning the model perform and calculate accuracy with all possible parameter given to it internally and give the best parameters
  which give the best performance.
- **Grid Search** and **Random Search** are two commonly used methods for Hyperparameter tunning.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
# creating dictionary --> key value pair of hyperparameters having key as parameter and values as its values
Lr_Hyperpara = {
    'C': [0.001, 0.01, 0.1, 1, 10, 1000],                 # Regularization strength values
    'solver': ['lbfgs', 'liblinear', 'saga'],             # Solver options
    'max_iter':[100,200,400,600,800,1000],                # maximum iteration
    'multi_class':['ovr','multinomial']                   # how to perform multiclass classification 
}

# training data on gridsearch cv for finding best parameters
Lr_grid = GridSearchCV(LogisticRegression(random_state=0),         # Estimator
                       param_grid=Lr_Hyperpara,      # param_grid----> hyperparametes(dictionary we created)
                       scoring='accuracy',            # scoring---> performance matrix to check performance
                       cv=3,                          # cv------> number of flods
                       verbose=3,                     # verbose=Controls the verbosity: the higher, the more messages.
                       n_jobs=-1                      # Number of jobs to run in parallel,-1 means using all processors.
                       )

# training data on gridsearch cv for finding best parameters
Lr_grid.fit(X_train,y_train)

print(f"Best Score: {Lr_grid.best_score_})")         # printing  best score
print(f"Best paramters: {Lr_grid.best_params_})")    # printing  best parameters
```

```python
# Train the final model with the best hyperparameters on the full training data
lr_hyp = LogisticRegression(C= 0.1, max_iter =  500, multi_class = 'multinomial', solver = 'lbfgs')
lr_hyp.fit(X_train, y_train)

# Evaluate the final model on the test data
y_hyp_lr = lr_hyp.predict(X_test)
acc_hyp_lr = accuracy_score(y_test, y_hyp_lr)
print("Accuracy of Logistic regression : ",acc_hyp_lr)
```

```python
print(classification_report(y_test,y_hyp_lr))
```
